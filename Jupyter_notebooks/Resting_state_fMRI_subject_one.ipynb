```python
import nibabel as nib
import nilearn as nil
from nilearn import plotting
from nilearn import image

data=nib.load("subject_rsfmri.nii")
print(image.load_img(data).shape)
data_first_volume=image.index_img(data,0)
plotting.plot_anat(data_first_volume,title="Subject_one",cut_coords=(0,0,0))
print("Working")
```
